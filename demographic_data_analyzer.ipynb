```python
import numpy as np
import pandas as pd
```

```python
y = y.rename(columns={'income': 'salary'})
```

```python
# Merge X and y into a single DataFrame
df = pd.concat([X, y], axis=1)
df
```

1. How many people of each race are represented in this dataset?

```python
def race_count(df):
    return df['race'].value_counts()

race_counts = race_count(df)
race_counts
```

2. What is the average age of men?

```python
def ave_age_men(df):
    return df[df['sex'] == 'Male']['age'].mean()

average_age_men = ave_age_men(df)
average_age_men 
```

3. What is the percentage of people who have a Bachelor's degree?

```python
def percentage_bachelors(df):
    return (df[df['education'] == 'Bachelors'].shape[0] / df.shape[0]) * 100

percent_bachelors = percentage_bachelors(df)
percent_bachelors
```

4. What percentage of people with advanced education (Bachelors, Masters, or Doctorate) make more than 50K?

```python
def higher_education_rich(df):
    higher_education = df[df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]
    return (higher_education[higher_education['salary'] == '>50K'].shape[0] / higher_education.shape[0]) * 100

higher_ed_rich = higher_education_rich(df)
higher_ed_rich
```

5. What percentage of people without advanced education make more than 50K?

```python
def lower_education_rich(df):
    lower_education = df[~df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]
    return (lower_education[lower_education['salary'] == '>50K'].shape[0] / lower_education.shape[0]) * 100

lower_ed_rich = lower_education_rich(df)
lower_ed_rich
```

6. What is the minimum number of hours a person works per week?

```python
def min_work_hours(df):
    return df['hours-per-week'].min()

min_hours = min_work_hours(df)
min_hours
```

7. What percentage of the people who work the minimum number of hours per week have a salary of more than 50K?

```python
def rich_percentage_min_hours(df):
    min_hours = min_work_hours(df)
    num_min_workers = df[df['hours-per-week'] == min_hours]
    return (num_min_workers[num_min_workers['salary'] == '>50K'].shape[0] / num_min_workers.shape[0]) * 100

rich_percentage = rich_percentage_min_hours(df)
rich_percentage
```

8. What country has the highest percentage of people that earn >50K and what is that percentage?

```python
def highest_earning_country(df):
    country_salary = df[df['salary'] == '>50K']['native-country'].value_counts()
    country_total = df['native-country'].value_counts()
    highest_earning_country = (country_salary / country_total * 100).idxmax()
    highest_earning_country_percentage = (country_salary / country_total * 100).max()
    return highest_earning_country, highest_earning_country_percentage

highest_country, highest_country_percent = highest_earning_country(df)
highest_country, highest_country_percent
```

9. Identify the most popular occupation for those who earn >50K in India.

```python
def top_in_occupation(df):
    india_occupations = df[(df['native-country'] == 'India') & (df['salary'] == '>50K')]['occupation']
    return india_occupations.value_counts().idxmax()

top_occupation = top_in_occupation(df)
top_occupation
```
